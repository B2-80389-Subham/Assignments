You own an ice cream business and you would like to create a model that could predict the daily revenue in dollars based on the outside air temperature (degC). Build Linear Regression Model and evaluate it with measuring different errors.

Dataset : IceCreamData.csv


### import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### read the data

```python
ice_data = pd.read_csv("IceCreamData.csv")
```

### Exploratory Data Analysis

```python
ice_data.info()
```

```python
ice_data.describe()
```

```python
sns.pairplot(ice_data)
```

```python
ice_data.corr()
```

### Data Processing

```python
# split the data in X(input) and y(output)
X = ice_data.drop(["Revenue"],axis = 1)

Y = ice_data["Revenue"]
```

```python
# split the data into train and test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3)
```

### Train the model

```python
from sklearn.linear_model import LinearRegression
ice_model = LinearRegression()

# fit the data/train the data
ice_model.fit(x_train,y_train)
```

### Evaluate the model

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
y_true  = y_test
y_pred = ice_model.predict(x_test)
```

```python
print(f"R2 Square = {r2_score(y_true,y_pred)}")
print(f"Mean Absolute Error = {mean_absolute_error(y_true,y_pred)}")
print(f"Mean Squared Error = {mean_squared_error(y_true,y_pred)}")
print(f"Root Mean Squared Error = {np.sqrt(r2_score(y_true,y_pred))}")
```

### Result Evaluation

```python
plt.scatter(ice_data["Temperature"],ice_data["Revenue"],c = ice_data["Temperature"])
plt.plot(X,ice_model.predict(X),color = "pink")
```
