```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
crime_df = pd.read_csv('crime_std.csv')
```

```python
crime_df.info()
```

```python
crime_df.describe()
```

```python
crime_df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

## Creating the model

```python
from scipy.cluster import hierarchy
```

```python
hmodel = hierarchy.linkage(crime_df, method='single')
result = hierarchy.dendrogram(hmodel)
```

```python
hmodel2 = hierarchy.linkage(crime_df, method='complete')
result2 = hierarchy.dendrogram(hmodel2)
```

```python
hmodel3 = hierarchy.linkage(crime_df, method='average')
result3 = hierarchy.dendrogram(hmodel3)
```

### Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
aggl_model = AgglomerativeClustering(n_clusters=5).fit(crime_df)
```

```python
plt.xlabel('UrbanPop')
plt.ylabel('Murder')
plt.title('UrbanPop vs Murder')

for i in range(5):
    plt.scatter(
        crime_df['UrbanPop'][aggl_model.labels_ == i],
        crime_df['Murder'][aggl_model.labels_ == i],
        label=f'Cluster {i}'
    )

plt.legend();
```
