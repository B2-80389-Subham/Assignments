# K-means clustering for crime_std data

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("crime_std.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
print(df.columns)
df.drop(["Unnamed: 0",], axis = 1, inplace = True)
print(df.columns)
```

```python
df.info()
```

```python
plt.scatter(df['Murder'],df['UrbanPop'], label = "Murder" , color = "red")
plt.xlabel("Murder")
plt.ylabel("UrbanPop")
plt.title("Murder vs UrbanPop")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans_model = KMeans(n_clusters = 3, random_state = 1)
kmeans_model.fit(df)
```

```python
kmeans_model.labels_
```

```python
kmeans_model.cluster_centers_
```

```python
df['Murder'][kmeans_model.labels_ == 0]
```

```python
plt.scatter(df['Murder'][kmeans_model.labels_ == 0],df['UrbanPop'][kmeans_model.labels_ == 0], color = "green", label = "cluster 0 ")
plt.scatter(df['Murder'][kmeans_model.labels_ == 1],df['UrbanPop'][kmeans_model.labels_ == 1], color = "pink", label = "cluster 1 ")
plt.scatter(df['Murder'][kmeans_model.labels_ == 2],df['UrbanPop'][kmeans_model.labels_ == 2], color = "red" , label = "cluster 2 ")

plt.title("Murder vs UrbanPop")
plt.xlabel("Murder")
plt.ylabel("UrbanPop")
plt.legend()
```
