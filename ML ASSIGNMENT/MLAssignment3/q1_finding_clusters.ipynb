```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("East_std.csv")
df
```

```python
print(df.columns)
df.drop(["Unnamed: 0",], axis = 1, inplace = True)
print(df.columns)
```

```python
df.info()
```

```python
plt.scatter(df['Qual_miles'],df['Days_since_enroll'], label = "Days_since_enroll" , color = "green")
plt.xlabel("Qual_miles")
plt.ylabel("Days_since_enroll")
plt.title("Qual_miles vs Days_since_enroll")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans_model = KMeans(n_clusters = 4, random_state = 12345)
kmeans_model.fit(df)
```

```python
kmeans_model.labels_
```

```python
kmeans_model.cluster_centers_
```

```python
df['Qual_miles'][kmeans_model.labels_ == 0]
```

```python
plt.scatter(df['Qual_miles'][kmeans_model.labels_ == 0],df['Balance'][kmeans_model.labels_ == 0], color = "green", label = "cluster 0 ")
plt.scatter(df['Qual_miles'][kmeans_model.labels_ == 1],df['Balance'][kmeans_model.labels_ == 1], color = "pink", label = "cluster 1 ")
plt.scatter(df['Qual_miles'][kmeans_model.labels_ == 2],df['Balance'][kmeans_model.labels_ == 2], color = "black" , label = "cluster 2 ")
plt.scatter(df['Qual_miles'][kmeans_model.labels_ == 3],df['Balance'][kmeans_model.labels_ == 3], color = "orange", label = "cluster 3 ")
plt.scatter(df['Qual_miles'][kmeans_model.labels_ == 4],df['Balance'][kmeans_model.labels_ == 4], color = "cyan", label = "cluster 4 ")

plt.title("Qual_miles vs Balance")
plt.xlabel("income")
plt.ylabel("spending_score")
plt.legend()
```
