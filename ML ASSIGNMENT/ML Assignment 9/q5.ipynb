```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SAT-GPA.csv")
```

```python
df.head()
```

```python
X = df.drop(["GPA"], axis =1)
Y = df['GPA']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y, train_size = 0.7 ,random_state=12345)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
plt.scatter(X, Y)
```

```python
model.predict(X)
plt.scatter(X,Y)
plt.plot(X,model.predict(X), color ="red",label= "regression line"  )
```

# polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
x_new_train = poly.fit_transform(x_train)
x_new_test = poly.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_new_train, y_train)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test, model.predict(x_new_test) ,color = "red")
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
r2 = r2_score(y_test,model.predict(x_new_test) )
mae = mean_absolute_error(y_test,model.predict(x_new_test))
mse = mean_squared_error(y_test,model.predict(x_new_test))
rmse = np.sqrt(mse)
```

```python
print(f"r2 score = {r2}")
```
