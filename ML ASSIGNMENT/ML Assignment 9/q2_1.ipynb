```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("NewspaperData.csv")
```

```python
df.head()
```

```python
df["Newspaper"].unique()
```

```python
X = df.drop(['Newspaper','sunday'], axis =1)
Y = df['sunday']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y, train_size = 0.7 ,random_state=12345)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
plt.scatter(df['daily'],df['sunday'])
```

```python
model.predict(X)
plt.scatter(df['daily'],df['sunday'])
plt.plot(X,model.predict(X), color ="red",label= "regression line"  )
```

# polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
x_new_train = poly.fit_transform(x_train)
x_new_test = poly.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_new_train, y_train)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test, model.predict(x_new_test) ,color = "red")
```
